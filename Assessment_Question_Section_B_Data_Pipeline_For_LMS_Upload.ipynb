```python
df=dfny.append(dfcg).append(dfws)
df
```

```python
df['Start Time'] = pd.to_datetime(df['Start Time'])

df['mostpopmth'] = pd.to_datetime(df['Start Time']).dt.month
#print("What is the most popular month for start time? ",df['mostpopmth'].mode())

df['mostpopdoy'] = pd.to_datetime(df['Start Time']).dt.day_of_week
#print("What is the most popular day of week (Monday, Tuesday, etc.) for start time? ",df['mostpopdoy'].mode())

df['mostpophod'] = pd.to_datetime(df['Start Time']).dt.hour
#print("What is the most popular hour of day for start time?",df['mostpophod'].mode())

df['mostpoptrip'] = df['Start Station']+df['End Station']

df.mode(axis=0)
```

```python
dfttd = df['Trip Duration'].sum()
dfatd = df['Trip Duration'].mean()
print("What is the total trip duration and average trip duration?", round(dfttd, 0), " ", round(dfatd,0))

print("What are the counts of each user type? \n")
print(df['User Type'].value_counts())
```
